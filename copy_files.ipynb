```python
import os
import shutil
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\Prayag\\Desktop\\2007_to_2013\\doc_names_links.csv')
df = df[:8605]
df
```

```python
origin_docs = list(df['doc_names'])
```

```python
origin_docs = [s + '.pdf' for s in origin_docs]
origin_docs
```

```python
# Providing the folder path
origin = r'C:\Users\Prayag\Desktop\finra_data\before_2013'
target = r'C:\Users\Prayag\Desktop\2007_to_2013'

# Fetching the list of all the files
# origin= os.listdir(origin)
origin_files = os.listdir(origin)
target_files = os.listdir(target)
# Fetching all the files to directory
for file_name in origin_docs:
    if file_name not in target_files:
        origin_address = origin + '\\' + file_name
        target_address = target + '\\' + file_name
        shutil.copy(origin_address, target_address)
        print("File copied", file_name)
```
